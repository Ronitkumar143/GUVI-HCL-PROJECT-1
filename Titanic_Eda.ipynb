```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/drive/MyDrive/titanic.csv")

print(df.head())
```

```python
# --- Countplot: Survived vs Not Survived ---
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=df, palette='viridis')
plt.title("Survival Count")
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.xlabel("Survival")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python

# --- Bar Plot: Survival by Gender and Class ---
plt.figure(figsize=(4,6))
sns.countplot(x='Sex', hue='Survived', data=df, palette='Set2')
plt.title("Survival by Gender")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# --- Pie Chart: Class Distribution ---
class_counts = df['Sex'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Set3"))
plt.title("Class Distribution")
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# --- Histogram: Age Distribution ---
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
